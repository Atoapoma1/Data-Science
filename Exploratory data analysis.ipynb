```python
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd

df = pd.read_csv("C:/Users/HP/DATA FOR DATA SCIENCE/mexico-real-estate-clean2.csv")

df.head()
```

```python
#Visualizing location data
fig = px.scatter_mapbox(
    df,
    lat="lat",
    lon="lon",
    center={"lat": 19.43, "lon": -99.13},
    width=600,
    height=600,
    hover_data=["price_usd"],
    mapbox_style = "open-street-map"
)

fig.show()
```

```python
#Checking how houses are distributed using histogram
plt.hist(df["area_m2"])
plt.xlabel("Area sq meters")
plt.ylabel("Frequency")
plt.title("Distribution of home sizes")
```

```python
# Use Matplotlib to create boxplot of "area_m2"
plt.boxplot(df["area_m2"], vert=False)
plt.xlabel("Area sq meters")
plt.title("Distribution of home sizes")
```

```python
# Use Matplotlib to create histogram of "price_usd"
plt.hist(df["price_usd"])
plt.xlabel("Price USD")
plt.ylabel("Frequency")
plt.title("Distribution of home prices")
```

```python
# Use Matplotlib to create boxplot of "price_usd"
plt.boxplot(df["price_usd"], vert=False)
plt.xlabel("Price USD")
plt.title("Distribution of Home Prices")
```
