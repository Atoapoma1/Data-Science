```python
import pandas as pd

# Load CSV file into DataFrame with specified encoding
df1 = pd.read_csv("C:/Users/HP/DATA FOR DATA SCIENCE/mexico-real-estate-1.csv", encoding='latin-1')
df2 = pd.read_csv("C:/Users/HP/DATA FOR DATA SCIENCE/mexico-real-estate-2.csv", encoding='latin-1')
df3 = pd.read_csv("C:/Users/HP/DATA FOR DATA SCIENCE/mexico-real-estate-3.csv", encoding='latin-1')


# Display the first few rows of the DataFrame
# df1["property_type"]
print("df1 type:", type(df1))
print("df1 shape:", df1.shape)
```

```python
# Print df1 shape
df1.shape
df2.shape
df1.shape

# Print df1 info
df1.info()

# Get output of df1 head
df1.head()
```

```python
#Remove NaN values
df1.dropna(inplace=True)
df1.info()
```

```python
df1["price_usd"].head()
```

```python
#cleaning price_usd for df3
df1["price_usd"] = df1["price_usd"].str.replace("$","").str.replace(",","").astype(float) 
df1["price_usd"].head()
```

```python
ato = "$123,567.00"
ato = float(ato.replace("$","").replace(",",""))
print(ato)
print(type(ato))
```

```python
#cleaning df2
df2.dropna(inplace=True)
df2.head()
```

```python
#We need to create a column called price_usd to convert price_mxn
#create a "price_usd"
df2["price_usd"] = (df2["price_mxn"] / 19).round(2)

# Drop "price_mxn" column from df2
df2.drop(columns=["price_mxn"],inplace=True)
df2.head()
```

```python
#cleaning df3
df3.info()
df3.dropna(inplace=True)
```

```python
df3.head()
```

```python
# Create "lat" and "lon" columns for df3
df3[["lat", "lon"]] = df3["lat-lon"].str.split(",", expand=True)
df3.head()
```

```python
#Using slpit method to create a state column and dropping "place_with_parent_names" and "lat-lon" columns from the DataFrame.
df3["state"] = df3["place_with_parent_names"].str.split("|", expand=True)[2]

df3.drop(columns=["place_with_parent_names", "lat-lon"], inplace=True)

df3.head()
```

```python
#Concatenating DF1, DF2 and DF3
df = pd.concat([df1, df2, df3])
df.head()
```

```python
#Saving df as a csv
df.to_csv("C:/Users/HP/DATA FOR DATA SCIENCE/mexico-real-estate-clean.csv", index=False)
```
